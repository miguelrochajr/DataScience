```python
import seaborn as sns
%matplotlib inline
tips = sns.load_dataset('tips')
tips.head()
```

```python
import numpy as np
```

```python
sns.barplot(x='sex', y='total_bill', data=tips, estimator=np.std)
```

```python
sns.countplot(x='sex',data=tips) #just is going to count the occurencies
```

```python
sns.boxplot(x='day',y='total_bill',data=tips)# hue='smoker')

# Understanding the Boxplots
# https://www.wellbeingatschool.org.nz/information-sheet/understanding-and-interpreting-box-plots
```

```python
sns.violinplot(x='day',y='total_bill', data=tips)
```

```python
sns.stripplot(x='day',y='total_bill', data=tips, jitter=True) #Like a Scatterplot
```

```python
sns.swarmplot(x='day',y='total_bill', data=tips, color='black') #Like a combinatin of a Violin Plot and a swarm plot
sns.violinplot(x='day',y='total_bill', data=tips)
```

```python
sns.factorplot(x='sex',y='total_bill',data=tips,kind='violin')
```
